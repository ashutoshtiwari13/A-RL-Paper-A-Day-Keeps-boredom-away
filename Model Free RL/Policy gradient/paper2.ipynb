## Paper : Trust Region Policy Optimization

### Main Focus

Demonstrate an iterative proceedure for optimizing policies with monotonic improvement. The authors propose a *policy gradient algorithm* called TRPO effective on non-linear policies such as neural networks.

Policy optimization can be classified into 3 broad categories,
1) Policy iteration - estimation , evaluation and improvement 

2) Policy gradient - estimator of the gradient of the total rewards

3) Derivative-free optimization- 
     - Cross-Entropy Method (CEM)
     - Covariance Matrix Adaptation( CAM) - very successful on continuous control problems 
    
Goal : To prove that minimizing a certain surrogate objective function guarantees policy improvement with non-trivial step sizes. Practically feasible algorithm called TRPO is implemented with two -variants *single path method*--- for model free setting and *vine method*--- only possible in simulation  
    

### Challenges/problems Identified

Previous model-free policy search algorithm had problems in learning complex polices, scalability and optimizing non-lieanr policies with tens and thousands of parameters.

### Sampling schemes used
- **Single-path method** :  The $s_{0} \sim p_{0}$ is sampled to collect sequence of states and then a trajectory is generated by simulating the policy $\pi_{old}$ parametrized over $\theta$

- **Vine method** :  The $s_{0} \sim p_{0}$ is sampled to collect sequence of states and a number of trajectories are generated . A subset of $\mathcal{N}$ states is selected along these trajectories, which are called "rollout sets". For each state in rollout set , $\mathcal{K}$ actions are sampled. For each of these K actions , Q-values are estimated.
